## Creating a word cloud where the color of each word is determined by its sentiment polarity. 

```python
from wordcloud import WordCloud, get_single_color_func
from textblob import TextBlob
import matplotlib.pyplot as plt

# Define a function to color words based on their polarity
def color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    sentiment = TextBlob(word).sentiment.polarity
    if sentiment > 0:
        return "green"  # Positive sentiment
    elif sentiment < 0:
        return "red"    # Negative sentiment
    else:
        return "grey"   # Neutral sentiment

# Sample text
sample_text = """
The morning started off beautifully, with the sun casting a warm glow through the window. Although the coffee tasted bitter, the gentle breeze and calm surroundings made it easy to relax. Later, the internet crashed unexpectedly, causing frustration and delays. Still, the team remained optimistic, and the meeting ended on a hopeful note. Some tasks were boring, but the overall progress was steady and encouraging.
"""

# Generate the word cloud
wordcloud = WordCloud(background_color="white", color_func=color_func).generate(sample_text)

# Display the word cloud using matplotlib
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
